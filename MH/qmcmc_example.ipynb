```python
from qmcmc import IsingMH
import numpy as np
import matplotlib.pyplot as plt
```

```python
s = 10
js_1d = np.random.normal(0, 1, s)
hs = np.random.normal(0, 1, s)
r = 5
t = 0.5

test_c = IsingMH(s, js_1d, hs, r, t, "classical")
test_q = IsingMH(s, js_1d, hs, r, t, "quantum")

_, _, _, a_c, rs_c = test_c.run_mcmc(100, 0)
_, _, _, a_q, rs_q = test_q.run_mcmc(100, 0)


plt.plot(-rs_c[0], label='Classical')
plt.plot(-rs_q[0], label='Quantum')
plt.xlabel("Iteration")
plt.ylabel("Energy")
plt.legend()
plt.show()
```
